```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cleaned_master=pd.read_csv('cleaned_master.csv')
```

```python
numeric_cols = cleaned_master.select_dtypes(include='number')
plt.figure(figsize=(10,8))
sns.heatmap(numeric_cols.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#ESG vs GDP, ESG vs SDI, Climate Risk vs ESG, etc.
sns.scatterplot(data=cleaned_master,x="GDP_2020",y="ESG_Final")
plt.title("Scatter plot GDP VS ESG")
plt.show()
sns.scatterplot(data=cleaned_master,x="SDI_2019",y="ESG_Final")
plt.title("Scatter plot SDI VS ESG")
plt.show()
sns.scatterplot(data=cleaned_master,x="Temperature",y="ESG_Final")
plt.title("Scatter plot Temperature VS ESG")
plt.show()
```
