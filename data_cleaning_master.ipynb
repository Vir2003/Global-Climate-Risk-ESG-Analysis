In this file we are doing all the neccessary cleaning and processing on master.csv

```python
import pandas as pd

final_master = pd.read_csv("final_master_test.csv")
print(final_master.info())  # Overview of columns and missing values
print(final_master.isna().sum()) 
```

```python
numeric_cols = final_master.select_dtypes(include=['float64', 'int64']).columns
final_master[numeric_cols] = final_master[numeric_cols].fillna(final_master[numeric_cols].median())
```

```python
categorical_cols = final_master.select_dtypes(include=['object']).columns
final_master[categorical_cols] = final_master[categorical_cols].fillna('Unknown')
```

```python
cols_to_fill = [
    "Access to clean fuels and technologies for cooking (% of population)",
    "Adjusted savings: natural resources depletion (% of GNI)",
    "Adjusted savings: net forest depletion (% of GNI)"
]

for col in cols_to_fill:
    final_master[col] = final_master[col].fillna(final_master[col].median())

final_master[cols_to_fill].isna().sum()
len(final_master)
```

```python
cols_to_drop = [
    "Access to clean fuels and technologies for cooking (% of population)",
    "Adjusted savings: natural resources depletion (% of GNI)",
    "Adjusted savings: net forest depletion (% of GNI)",
    "CO2 emissions (metric tons per capita)",
    "Renewable energy consumption (% of total final energy consumption)"
]

final_master = final_master.drop(columns=cols_to_drop, errors='ignore')
print("Dropped null ESG columns. Shape after clean-up:", final_master.shape)
```

```python
import numpy as np

def convert_population(val):
    if isinstance(val, str):
        val = val.strip().upper()
        if val.endswith('B'):
            return float(val[:-1]) * 1e9
        elif val.endswith('M'):
            return float(val[:-1]) * 1e6
        elif val.endswith('K'):
            return float(val[:-1]) * 1e3
        else:
            try:
                return float(val)
            except:
                return np.nan
    return val

final_master['POP_2020'] = final_master['POP_2020'].apply(convert_population)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize GDP_per_capita and Population before combining
scaler = MinMaxScaler()
final_master[['GDP_per_capita_norm', 'POP_norm']] = scaler.fit_transform(
    final_master[['GDP_2020', 'POP_2020']]
)

final_master['Economic_Index'] = final_master[['GDP_per_capita_norm', 'POP_norm']].mean(axis=1)
```

```python
final_master['Sustainability_Index'] = final_master[['ESG_Final', 'SDI_2019']].mean(axis=1)
final_master['Climate_Risk_Index'] = final_master['Temperature']
```
